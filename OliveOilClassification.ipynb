```python
from sklearn import metrics
from scipy.io import arff
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix, precision_score, recall_score, f1_score
import numpy as np

# Load the training data
with open('OliveOil_TRAIN.arff') as f:
    data, meta = arff.loadarff(f)
df_train = pd.DataFrame(data)
X_train = df_train.iloc[:,:-1].values
y_train = df_train.iloc[:,-1].values

# Load the test data
with open('OliveOil_TEST.arff') as f:
    data, meta = arff.loadarff(f)
df_test = pd.DataFrame(data)
X_test = df_test.iloc[:,:-1].values
y_test = df_test.iloc[:,-1].values

print(df_train.shape)
print(df_test.shape)

# Display all the rows of training data
df_train.head(30)
df_train.describe()
print( df_train['target'].value_counts())

# Encode the target variable
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

# Plot the variation of every feature for the different samples of training data
for i in range(X_train.shape[1]):
    plt.plot(X_train[:, i])
plt.title('Training data Features Variation')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.show()

# Plot the variation of every feature for the different samples of testing data
for i in range(X_test.shape[1]):
    plt.plot(X_test[:, i])
plt.title('Testing data Features Variation')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.show()


#--------------------------------- Nave Bayesian Classifier------------------------------------------
from sklearn.naive_bayes import GaussianNB

# Train the model
nb = GaussianNB()
nb.fit(X_train, y_train)

# Make predictions on the test data
y_pred = nb.predict(X_test)
y_predt = nb.predict(X_train)

# Evaluate the performance of the model
accuracy = accuracy_score(y_train, y_predt)
cm = confusion_matrix(y_train, y_predt)
prec = precision_score(y_train, y_predt, average='macro')
recall = recall_score(y_train, y_predt, average='macro')
f1 = f1_score(y_train, y_predt, average='macro')
clr= metrics.classification_report(y_train, y_predt)

print("Nave Bayesian Classifier:")
print("Train report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
clr= metrics.classification_report(y_test, y_pred)

print("Test report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

#---------------------------------K-Nearest Neighbors (KNN)-----------------------------------
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for the number of neighbors (k)
param_grid = {'n_neighbors': [3, 5, 7, 9, 11]}

# Create a KNN classifier
knn = KNeighborsClassifier()

# Create a grid search object with cross-validation
grid_search = GridSearchCV(knn, param_grid=param_grid, cv=5)

# Fit the grid search object to the training data
grid_search.fit(X_train, y_train)

# Print the best value of k and the corresponding accuracy score
print("Best k:", grid_search.best_params_['n_neighbors'])
print("Accuracy:", grid_search.best_score_)

# Use the best classifier to predict the target values for the test data
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(X_test)
y_predt = best_knn.predict(X_train)

# Evaluate the performance of the model
accuracy = accuracy_score(y_train, y_predt)
cm = confusion_matrix(y_train, y_predt)
prec = precision_score(y_train, y_predt, average='macro')
recall = recall_score(y_train, y_predt, average='macro')
f1 = f1_score(y_train, y_predt, average='macro')
clr= metrics.classification_report(y_train, y_predt)

print("K-Nearest Neighbors Classifier:")
print("Train report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
clr= metrics.classification_report(y_test, y_pred)

print("Test report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

#------------------------------------Decision Tree-----------------------------------------
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
sc.fit(X_test)
X_test = sc.transform(X_test)

# fit the model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# make predictions on the test set
y_pred = dt.predict(X_test)
y_predt = dt.predict(X_train)

# Evaluate the performance of the model
accuracy = accuracy_score(y_train, y_predt)
cm = confusion_matrix(y_train, y_predt)
prec = precision_score(y_train, y_predt, average='macro')
recall = recall_score(y_train, y_predt, average='macro')
f1 = f1_score(y_train, y_predt, average='macro')
clr= metrics.classification_report(y_train, y_predt)

print("Decision Tree Classifier:")
print("Train report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
clr= metrics.classification_report(y_test, y_pred)

print("Test report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

#------------------------------Linear Discriminate Analysis (LDA)---------------------------------
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Create the LDA model
lda = LinearDiscriminantAnalysis()

# Train the model on the training data
lda.fit(X_train, y_train)

# Make predictions on the test data
y_pred = lda.predict(X_test)
y_predt = lda.predict(X_train)

# Evaluate the performance of the model
accuracy = accuracy_score(y_train, y_predt)
cm = confusion_matrix(y_train, y_predt)
prec = precision_score(y_train, y_predt, average='macro')
recall = recall_score(y_train, y_predt, average='macro')
f1 = f1_score(y_train, y_predt, average='macro')
clr= metrics.classification_report(y_train, y_predt)

print("Linear Discriminate Analysis Classifier:")
print("Train report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
clr= metrics.classification_report(y_test, y_pred)

print("Test report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

from sklearn import tree
tree.export_graphviz(dt, out_file='tree.dot')
from io import StringIO
import pydot
out_data = StringIO()
# Exports the trained Decision Tree dt in Graphviz format as a .dot file named tree.dot
tree.export_graphviz(dt, out_file=out_data, feature_names=df_train.iloc[1,:-1].values, class_names=dt.classes_.astype(int).astype(str), filled=True, rounded=True,special_characters=True,
node_ids=1,)
# Read the graph data from out_data and create a pydot graph object
graph = pydot.graph_from_dot_data(out_data.getvalue())
# Save the graph as a PDF file named oliveOil.pdf
graph[0].write_pdf("oliveOil.pdf") 

#---------------------------------Artificial Neural Networks (ANN)----------------------------
from sklearn.neural_network import MLPClassifier

# Initialize ANN classifier
mlp = MLPClassifier(activation='logistic', alpha=0.0001, batch_size='auto', beta_1=0.9, beta_2=0.999, early_stopping=False, epsilon=1e-08, hidden_layer_sizes=(100, 100, 100), learning_rate='constant',learning_rate_init=0.001, max_iter=100, momentum=0.9,nesterovs_momentum=True, power_t=0.5, random_state=None,shuffle=True, solver='adam', tol=0.0001, validation_fraction=0.1, verbose=False, warm_start=False)

# Train the classifier with the traning data
mlp.fit(X_train,y_train)
#The MLPClassifier is a multilayer perceptron neural network that can be used for classification tasks
# There are three hidden layers, each with 30 neurons; the learning rate is set to 0.001
# The activation function used in the hidden layers is the sigmoid function ('logistic')
print("Training set score: %f" % mlp.score(X_train, y_train))
print("Test set score: %f" % mlp.score(X_test, y_test))

# Make predictions on the testing data
y_pred = mlp.predict(X_test)
y_predt = mlp.predict(X_train)

# Evaluate the performance of the model
accuracy = accuracy_score(y_train, y_predt)
cm = confusion_matrix(y_train, y_predt)
prec = precision_score(y_train, y_predt, average='macro')
recall = recall_score(y_train, y_predt, average='macro')
f1 = f1_score(y_train, y_predt, average='macro')
clr= metrics.classification_report(y_train, y_predt)

print("Artificial Neural Networks Classifier:")
print("Train report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
clr= metrics.classification_report(y_test, y_pred)

print("Test report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

#--------------------------------------Support Vector Machine (SVM)--------------------------------------
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC

# Define the hyperparameter grid
param_grid = {'C': np.logspace(-3, 3, 7),'gamma': np.logspace(-3, 3, 7),'kernel': ['linear', 'rbf', 'poly']}

# Create the SVM model
svm = SVC()

# Create a randomized search object
random_search = RandomizedSearchCV(estimator=svm, param_distributions=param_grid, n_iter=100, cv=5, random_state=42)

# Fit the randomized search object to the data
random_search.fit(X_train, y_train)

print("Support Vector Machine Classifier:")
# Print the best parameters and score
print("Best parameters:", random_search.best_params_)
print("Best score:", random_search.best_score_)

# Use the best estimator to predict on the test set
y_pred = random_search.best_estimator_.predict(X_test)
y_predt = random_search.best_estimator_.predict(X_train)

# Evaluate the performance of the model
accuracy = accuracy_score(y_train, y_predt)
cm = confusion_matrix(y_train, y_predt)
prec = precision_score(y_train, y_predt, average='macro')
recall = recall_score(y_train, y_predt, average='macro')
f1 = f1_score(y_train, y_predt, average='macro')
clr= metrics.classification_report(y_train, y_predt)

print("Train report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
clr= metrics.classification_report(y_test, y_pred)

print("Test report:")
print("Accuracy:", accuracy*100," %")
print(f"Precision: {prec*100} % ")
print(f"Recall: {recall*100} %")
print(f"F1-score: {f1*100} %")
print("Confusion matrix:\n", cm)
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, cmap="plasma",  annot=True, linewidths=2, linecolor="white", cbar=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
print (f"Classification report :", clr)
```
